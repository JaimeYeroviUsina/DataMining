Importamos el dataset a evaluar y lo presentamos antes de preprocesarlo

```python
data = pd.read_csv("/kaggle/input/beijing-multisite-airquality-data-set/PRSA_Data_Nongzhanguan_20130301-20170228.csv")

data
```

```python
data = data.dropna()

data.isnull().any().any()
```

```python
from sklearn.model_selection import train_test_split

X = data[['SO2', 'NO2', 'PM10']]
y = data['day']

#X1 = data[['CO', 'O3']]
#y1 = data['hour']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=42)
```

```python
#from sklearn.svm import SVC
from sklearn import tree
#from sklearn import linear_model

#model = SVC()
model = tree.DecisionTreeClassifier()
#model = tree.DecisionTreeRegressor()
#model = linear_model.LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#model.fit(X1_train, y1_train)
#y1_pred = model.predict(X1_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='micro')
recall = recall_score(y_test, y_pred, average='micro')
f1 = f1_score(y_test, y_pred, average='micro')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.model_selection import cross_val_score

# Example of cross-validation using 5 folds
scores = cross_val_score(model, X, y, cv=5)
print("Cross-Validation Scores:", scores)
print("Average Accuracy:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

# Example of hyperparameter tuning using grid search
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
print('Best parameters', best_params)
```
