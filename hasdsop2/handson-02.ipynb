```r
#Create a new data set by imputing the mean in all the columns which have double type values.

data <- read.csv("/kaggle/input/ucmachinelearning/imports-85.data")

df <- data.frame(data)

df[df == "?"] <- NA

double <- sapply(df, is.double)

for (col in names(df)[double]) {
 mean_value <- mean(df[[col]], na.rm = TRUE)
 df[[col]][is.na(df[[col]])] <- mean_value
}
head (mean_value,10)

df <- replace(df, is.na(df), mean_value)

df
```

```r
#Create a new data set by imputing the mode in all the columns which have integer type values.

data <- read.csv("/kaggle/input/ucmachinelearning/imports-85.data")

df <- data.frame(data)

df[df == "?"] <- NA

integer <- sapply(df, is.integer)

mode <- function(column) {
  mode_value <- as.numeric(names(table(column))[table(column) == max(table(column))])
  column[is.na(column)] <- mode_value
  return(column)
}

df <- replace(df, is.na(df), mode_value)

df
```

```r
#Create a new data set by imputing the most frequent value to the column ”nDoors”.

data <- read.csv("/kaggle/input/ucmachinelearning/imports-85.data")

df <- data.frame(data)

most_frequent <- names(sort(table(df$NUM.OF.DOORS), decreasing = TRUE))[1]

data_ndoors <- replace(df, is.na(df), most_frequent)

data_ndoors
```
